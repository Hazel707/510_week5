## Calculating The Diff in a Series

```python
# formula for calculating the delta between list values
sample_xa = list(range(1, 12, 2))
print(f'The list is {sample_xa}')
total = 0 # initialize total at 0
ceiling = len(sample_xa) - 1 # declare the ceiling which is last index position
for i, v in enumerate(sample_xa):
    if i < ceiling:
        diff = sample_xa[i + 1] - sample_xa[i]
        total += diff
print(total)
```
